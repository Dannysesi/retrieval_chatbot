```python
import pandas as pd
import json

def excel_to_json_intent(excel_file, output_file):
    # Read the Excel file
    df = pd.read_excel(excel_file)

    # Create a list to store intents
    intents = []

    # Iterate through each row of the dataframe
    for index, row in df.iterrows():
        # Split the symptoms into lists based on commas
        english_patterns = [symptom.strip() for symptom in row["Symptoms"].split(',')]
        yoruba_patterns = [symptom.strip() for symptom in row["Awọn aami aisan"].split(',')]

        # Create two separate intents: one for English and one for Yoruba
        english_intent = {
            "tag": row["English"],  # English disease name as tag
            "patterns": english_patterns,  # List of English patterns
            "responses": [row["English"]]  # Response in English
        }

        yoruba_intent = {
            "tag": row["Yoruba"],  # Yoruba disease name as tag
            "patterns": yoruba_patterns,  # List of Yoruba patterns
            "responses": [row["Yoruba"]]  # Response in Yoruba
        }

        # Append both intents to the list
        intents.append(english_intent)
        intents.append(yoruba_intent)

    # Add greetings to the intents
    intents.extend([
        {
            "tag": "greeting",
            "patterns": ["hi", "hello", "good morning", "good afternoon", "good evening", "kí lo", "báwo ni"],
            "responses": ["Hello", "Kí lo"]
        },
        {
            "tag": "goodbye",
            "patterns": ["goodbye", "bye", "see you later", "sábà", "àdánù"],
            "responses": ["Goodbye", "Sábà"]
        }
    ])

    # Write the intents to a JSON file
    with open(output_file, "w", encoding='utf-8') as f:
        json.dump(intents, f, indent=4)

# Replace with your Excel file and output file paths
excel_file = "cleaned_disease_data.xlsx"
output_file = "disease_intents1.json"

# Call the function
excel_to_json_intent(excel_file, output_file)
```
